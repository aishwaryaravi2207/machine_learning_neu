```python
import matplotlib.pyplot as plt

# Function to include mean along with min, max, median, and std. dev for each selected variable across all years
def plot_stats_for_selected_vars_with_mean(dfs, years, variables, stats_to_plot=['min', 'max', '50%', 'std', 'mean']):
    # Dictionary to store summary statistics for each year
    summary_stats = {stat: {var: [] for var in variables} for stat in stats_to_plot}
    
    # Loop through each dataset and extract the required statistics for each variable
    for df in dfs:
        stats = df[variables].describe()
        for stat in stats_to_plot:
            for var in variables:
                if stat == 'mean':
                    summary_stats[stat][var].append(df[var].mean())  # Get mean separately
                else:
                    summary_stats[stat][var].append(stats.loc[stat, var])
    
    # Create a separate plot for each variable's statistics
    for var in variables:
        plt.figure(figsize=(10, 6))
        for stat in stats_to_plot:
            plt.plot(years, summary_stats[stat][var], marker='o', label=stat)
        plt.xlabel('Year')
        plt.ylabel(f'{var} Value')
        plt.title(f'{var} Statistics Comparison Across Years')
        plt.legend()
        plt.grid(True)
        plt.show()

# List of datasets and corresponding years
datasets = [df_2015, df_2016, df_2017, df_2018, df_2019, df_2020, df_2021, df_2022, df_2023, df_2024]
years = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024]

# List of specified continuous variables
selected_variables = [
    'Test Veh Displacement (L)', 'Axle Ratio', 'N/V Ratio', 
    'CO2 (g/mi)', 'NOx (g/mi)', 'Set Coef A (lbf)', 
    'Target Coef A (lbf)', 'DT-Inertia Work Ratio Rating'
]

# Plot separate graphs for min, max, median, std. dev, and mean for each selected variable
plot_stats_for_selected_vars_with_mean(datasets, years, selected_variables)
```

```python
# Function to display only a few outliers to avoid truncation
def identify_and_display_outliers_iqr(dfs, variables, max_outliers=5):
    outliers = {var: [] for var in variables}
    
    # Loop through each dataset to identify outliers for the selected variables
    for year, df in zip(years, dfs):
        for var in variables:
            # Calculate Q1 (25th percentile) and Q3 (75th percentile)
            Q1 = df[var].quantile(0.25)
            Q3 = df[var].quantile(0.75)
            IQR = Q3 - Q1  # Interquartile range
            
            # Determine outliers: Data points outside 1.5 * IQR from Q1 and Q3
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            outliers_data = df[(df[var] < lower_bound) | (df[var] > upper_bound)][var]
            
            # Store a limited number of outliers and the corresponding year
            outliers[var].append((year, outliers_data.values[:max_outliers]))
    
    # Display the outliers with a limited number
    for var, data in outliers.items():
        print(f"\nOutliers for {var}:")
        for year, outliers_list in data:
            print(f"Year {year}: {outliers_list}")

# List of datasets and corresponding years
datasets = [df_2015, df_2016, df_2017, df_2018, df_2019, df_2020, df_2021, df_2022, df_2023, df_2024]
years = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024]

# List of specified continuous variables
selected_variables = [
    'Test Veh Displacement (L)', 'Axle Ratio', 'N/V Ratio', 
    'CO2 (g/mi)', 'NOx (g/mi)', 'Set Coef A (lbf)', 
    'Target Coef A (lbf)', 'DT-Inertia Work Ratio Rating'
]

# Apply the updated function to identify and display a few outliers for the selected variables
identify_and_display_outliers_iqr(datasets, selected_variables, max_outliers=5)
```

```python
# Function to calculate and drop outliers using IQR method and save back to original datasets
def drop_outliers_iqr_inplace(dfs, variables):
    for i, df in enumerate(dfs):
        for var in variables:
            # Calculate Q1 (25th percentile) and Q3 (75th percentile)
            Q1 = df[var].quantile(0.25)
            Q3 = df[var].quantile(0.75)
            IQR = Q3 - Q1  # Interquartile range
            
            # Determine outliers: Data points outside 1.5 * IQR from Q1 and Q3
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            
            # Drop outliers by keeping only data within the bounds
            dfs[i] = dfs[i][(dfs[i][var] >= lower_bound) & (dfs[i][var] <= upper_bound)]
    
    return dfs  # The original datasets are modified in-place

# List of standardized continuous variable names (matching lowercase and underscores)
selected_variables = [
    'Test Veh Displacement (L)', 'Axle Ratio', 'N/V Ratio', 
    'CO2 (g/mi)', 'NOx (g/mi)', 'Set Coef A (lbf)', 
    'Target Coef A (lbf)', 'DT-Inertia Work Ratio Rating'
]

# Assuming we have the list of datasets for each year
datasets = [df_2015, df_2016, df_2017, df_2018, df_2019, df_2020, df_2021, df_2022, df_2023, df_2024]

# Apply the drop_outliers_iqr_inplace function to modify the original datasets
datasets = drop_outliers_iqr_inplace(datasets, selected_variables)

# Example: Display the first few rows of the first dataset (2015) after outlier removal
print(datasets[0].head())  # Show the dataset after dropping outliers
```

```python
# Function to identify missing values in each dataset and display the columns with the most missing data
def check_missing_values(dfs, years):
    # Dictionary to store missing values count for each year
    missing_data_summary = {}
    
    # Loop through each dataset to identify missing values
    for df, year in zip(dfs, years):
        # Calculate the number of missing values for each column
        missing_counts = df.isnull().sum()
        
        # Filter columns with missing values
        columns_with_missing = missing_counts[missing_counts > 0]
        
        # Store the result if there are missing values
        if not columns_with_missing.empty:
            missing_data_summary[year] = columns_with_missing.sort_values(ascending=False)
    
    return missing_data_summary

# List of datasets and corresponding years
datasets = [df_2015, df_2016, df_2017, df_2018, df_2019, df_2020, df_2021, df_2022, df_2023, df_2024]
years = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024]

# Check for missing values in each dataset and display the columns with the most missing data
missing_values_summary = check_missing_values(datasets, years)

# Display the summary of missing data
for year, missing_data in missing_values_summary.items():
    print(f"\nMissing Data in {year} Dataset:")
    print(missing_data)
```

## Handling missing data
1. Remove Missing Data:
Drop Rows: Remove rows that contain missing values.
Drop Columns: Remove columns that have a significant portion of missing data.
2. Imputation (Fill Missing Data):
Fill with Constant: Replace missing values with a fixed value (e.g., 0).
Fill with Mean/Median/Mode: For numerical data, replace missing values with the column's mean, median, or mode.
Forward/Backward Fill: For time-series data, fill missing values with the previous or next values.
3. Predictive Imputation:
Machine Learning: Use algorithms (like KNN) to predict and fill missing values.
4. Flag Missing Data:
Create Indicator Column: Track missing values by creating a binary column indicating missing entries.

## 3. Comparing Column Names Across Years

```python
# Function to compare column names across datasets and check for inconsistencies
def compare_column_names(dfs, years):
    # Store column names for each dataset
    column_names = {}
    
    # Loop through each dataset to get the column names
    for df, year in zip(dfs, years):
        column_names[year] = set(df.columns)
    
    # Compare the column names across datasets
    all_columns = set.union(*column_names.values())
    differences = {}
    
    # Check for missing or extra columns in each year
    for year, cols in column_names.items():
        missing_columns = all_columns - cols
        extra_columns = cols - all_columns
        if missing_columns or extra_columns:
            differences[year] = {
                "Missing Columns": missing_columns,
                "Extra Columns": extra_columns
            }
    
    return differences

# List of datasets and corresponding years
datasets = [df_2015, df_2016, df_2017, df_2018, df_2019, df_2020, df_2021, df_2022, df_2023, df_2024]
years = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024]

# Compare the column names across datasets
column_name_differences = compare_column_names(datasets, years)

# Display the differences in column names or report no differences
if column_name_differences:
    for year, diff in column_name_differences.items():
        print(f"\nColumn Name Differences in {year} Dataset:")
        print(f"Missing Columns: {diff['Missing Columns']}")
        print(f"Extra Columns: {diff['Extra Columns']}")
else:
    print("No column name differences found across datasets.")
```

```python
import pandas as pd

# Function to standardize column names across datasets
def standardize_column_names(dfs):
    # Step 1: Normalize column names (convert to lowercase and replace spaces with underscores)
    standardized_dfs = []
    for df in dfs:
        df.columns = df.columns.str.lower().str.replace(' ', '_')
        standardized_dfs.append(df)
    
    # Step 2: Identify all unique columns across datasets
    all_columns = set().union(*[set(df.columns) for df in standardized_dfs])
    
    # # Step 3: Add missing columns to each dataset and fill with NaN
    # for df in standardized_dfs:
    #     missing_cols = all_columns - set(df.columns)
    #     for col in missing_cols:
    #         df[col] = pd.NA  # Add missing columns filled with NaN
    
    # Step 4: Reorder columns consistently across all datasets
    standardized_dfs = [df[sorted(all_columns)] for df in standardized_dfs]
    
    return standardized_dfs

# List of datasets (for example: df_2015, df_2016, ...)
datasets = [df_2015, df_2016, df_2017, df_2018, df_2019, df_2020, df_2021, df_2022, df_2023, df_2024]

# Standardize column names across datasets
standardized_datasets = standardize_column_names(datasets)

# Example of stacking the standardized datasets into one dataframe
final_combined_df = pd.concat(standardized_datasets, ignore_index=True)

# Display the combined dataframe
print(final_combined_df.head())
```

```python
import pandas as pd

# Function to display sorted correlation values for all years datasets
def display_sorted_correlations_for_all_years(datasets, years, threshold=0.5):
    for df, year in zip(datasets, years):
        print(f"\nSorted Correlations for Model Year {year}:")
        
        # Select only numeric columns
        numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
        
        # Calculate the correlation matrix
        corr_matrix = df[numeric_cols].corr()
        
        # Handle missing or infinite values by filling with zeros
        corr_matrix = corr_matrix.replace([pd.NA, np.inf, -np.inf], 0)
        
        # Get the correlations as a stack and sort them
        corr_pairs = corr_matrix.unstack().reset_index()
        corr_pairs.columns = ['Variable 1', 'Variable 2', 'Correlation']
        
        # Filter out self-correlations and values below the threshold
        filtered_corr_pairs = corr_pairs[(corr_pairs['Variable 1'] != corr_pairs['Variable 2']) & 
                                         (abs(corr_pairs['Correlation']) >= threshold)]
        
        # Sort the correlations by absolute value, in descending order
        sorted_corr_pairs = filtered_corr_pairs.sort_values(by='Correlation', ascending=False, key=abs)
        
        # Display the top 20 sorted correlations for the given year
        print(sorted_corr_pairs.head(20))

# Assuming we have the list of datasets for each year and corresponding years
datasets = [df_2015, df_2016, df_2017, df_2018, df_2019, df_2020, df_2021, df_2022, df_2023, df_2024]
years = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024]

# Display sorted correlations for all years datasets
display_sorted_correlations_for_all_years(datasets, years, threshold=0.5)




```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Function to calculate correlations and plot heatmap for emission-related variables for each year
def plot_emission_correlations(datasets, years, variables):
    for df, year in zip(datasets, years):
        print(f"\nCorrelation heatmap for model year: {year}")
        # Select only the specified emission variables
        df_emissions = df[variables]
        
        # Calculate the correlation matrix for these emission variables
        corr_matrix = df_emissions.corr()
        
        # Plot the heatmap for the correlation matrix
        plt.figure(figsize=(8, 6))
        sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
        plt.title(f'Emission Correlations for Model Year {year}')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()

# List of emissions-related variables
emission_variables = ['co2_(g/mi)', 'nox_(g/mi)', 'thc_(g/mi)', 'co_(g/mi)', 'pm_(g/mi)']

# Assuming we have the list of datasets for each year and corresponding years
datasets = [df_2015, df_2016, df_2017, df_2018, df_2019, df_2020, df_2021, df_2022, df_2023, df_2024]
years = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024]

# Plot the correlation heatmap for emission-related variables across all years
plot_emission_correlations(datasets, years, emission_variables)

```
